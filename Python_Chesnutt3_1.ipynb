##Creating the First Subset
Now that you have all of the data, you can begin to create a smaller subset.
- For this subset, we want Python to only recognize data from NC.
Therefore we need to create a command that isolates every instance of "State" as "NC".
>The statement should contain `df["State"] == NC`

```python
df["State"]
```

```python
df["State"]=="NC"
```

###Now that Python recognizes NC as `"State"`, we can create a table that includes the columns and rows that we need.
- We will be looking at Adult obsesity and Unemployment in NC. Therefore, we need to select those two columns and then select a certain number of rows as shown below.

>The rows selected are labeled as `[3295:3338]`

>The columns are labeled as `["State","Adult Obesity","Year", "Unemployment"]`

```python
df[["State","Adult obesity","Year","Unemployment"]][3295:3338]
```

###Naming your Newly Created Subset
- Now that we have pulled parts of the data from the larger set of data and created a new table, we came name the subset.
- It is easiest to name it something that we can easily remember. Therefore, I chose to name my subset, `"Subset"`
>1. Put the name you have chosen on the left side of the `=`.
>2. Use the function "loc" to read your newly complied data set.
>3. Inside of the brackets, first include your columns `3295:3338`. Then add a comma and include your rows.
>4. Be sure to add `.copy()` to the end so you do not get an error.

```python
subset = df.loc[3295:3338, ["State","Adult obesity","Year","Unemployment"]].copy()
```

```python
subset
```
